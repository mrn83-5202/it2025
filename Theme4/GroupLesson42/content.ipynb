üîπ **–¢–∏–ø–∏ API:**  
‚úî **REST API** ‚Äì –Ω–∞–π–ø–æ—à–∏—Ä–µ–Ω—ñ—à–∏–π —Å—Ç–∞–Ω–¥–∞—Ä—Ç –æ–±–º—ñ–Ω—É –¥–∞–Ω–∏–º–∏ —á–µ—Ä–µ–∑ HTTP.  
‚úî **GraphQL API** ‚Äì –¥–æ–∑–≤–æ–ª—è—î –∫–ª—ñ—î–Ω—Ç—É –æ—Ç—Ä–∏–º—É–≤–∞—Ç–∏ –ª–∏—à–µ –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ.  
‚úî **gRPC API** ‚Äì –≤–∏—Å–æ–∫–æ–ø—Ä–æ–¥—É–∫—Ç–∏–≤–Ω–∏–π –±—ñ–Ω–∞—Ä–Ω–∏–π –ø—Ä–æ—Ç–æ–∫–æ–ª –æ–±–º—ñ–Ω—É.  

üìå **–ü—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è REST API —É Flask:**

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/status")
async def status():
    return {"service": "active", "version": "1.0"}

if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
