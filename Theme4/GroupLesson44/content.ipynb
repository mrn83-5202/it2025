üîπ **–ö–ª—é—á–æ–≤—ñ —Ç–µ—Ä–º—ñ–Ω–∏:**
- **–ê—Å–ø–µ–∫—Ç (Aspect):** –∫–æ–¥, —â–æ —Ä–µ–∞–ª—ñ–∑—É—î –Ω–µ—Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—å–Ω—É –ø–æ–≤–µ–¥—ñ–Ω–∫—É.  
- **–ü–æ—ñ–Ω—Ç–∫–∞—Ç (Pointcut):** –º—ñ—Å—Ü–µ –≤ –∫–æ–¥—ñ, –¥–µ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∞—Å–ø–µ–∫—Ç.  
- **Advice:** –¥—ñ—è, —è–∫–∞ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –ø—ñ–¥ —á–∞—Å pointcut (before, after, around).

### **3.2. –ê–û–ü —É –¥—ñ—ó (–Ω–∞ –ø—Ä–∏–∫–ª–∞–¥—ñ Python)**  
üìå **–†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –ª–æ–≥—É–≤–∞–Ω–Ω—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä–∞ ‚Äî –∞–Ω–∞–ª–æ–≥ –∞—Å–ø–µ–∫—Ç–∞**
```python
def log_request(func):
    def wrapper(*args, **kwargs):
        print(f"[LOG] Calling {func.__name__} with {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"[LOG] Finished {func.__name__}")
        return result
    return wrapper

@log_request
def process_data(data):
    return data.upper()

process_data("input")
```
