## **2. –û—Å–Ω–æ–≤–∏ –∞–Ω–∞–ª—ñ–∑—É –¥–∞–Ω–∏—Ö –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –∑–≥–æ—Ä—Ç–∫–æ–≤–∏—Ö –Ω–µ–π—Ä–æ–Ω–Ω–∏—Ö –º–µ—Ä–µ–∂ (40 —Ö–≤)**  

### **2.1. –©–æ —Ç–∞–∫–µ –∑–≥–æ—Ä—Ç–∫–æ–≤—ñ –Ω–µ–π—Ä–æ–Ω–Ω—ñ –º–µ—Ä–µ–∂—ñ (CNN)?**  
üìå **–ó–≥–æ—Ä—Ç–∫–æ–≤—ñ –Ω–µ–π—Ä–æ–Ω–Ω—ñ –º–µ—Ä–µ–∂—ñ (Convolutional Neural Networks, CNN)** ‚Äì —Ü–µ –∞—Ä—Ö—ñ—Ç–µ–∫—Ç—É—Ä–∞ –≥–ª–∏–±–æ–∫–æ–≥–æ –Ω–∞–≤—á–∞–Ω–Ω—è, —â–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –æ–±—Ä–æ–±–∫–∏ –∑–æ–±—Ä–∞–∂–µ–Ω—å.  

üîπ **–û—Å–Ω–æ–≤–Ω—ñ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∏ CNN:**  
- **–ó–≥–æ—Ä—Ç–∫–æ–≤–∏–π —à–∞—Ä (Convolution Layer)** ‚Äì –≤–∏–¥—ñ–ª–µ–Ω–Ω—è –æ–∑–Ω–∞–∫ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è.  
- **–®–∞—Ä —Å—É–±–¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü—ñ—ó (Pooling Layer)** ‚Äì –∑–º–µ–Ω—à–µ–Ω–Ω—è —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ –¥–∞–Ω–∏—Ö.  
- **–ü–æ–≤–Ω–æ–∑–≤‚Äô—è–∑–Ω–∏–π —à–∞—Ä (Fully Connected Layer)** ‚Äì –ø—Ä–∏–π–Ω—è—Ç—Ç—è —Ä—ñ—à–µ–Ω–Ω—è.  

üìå **–ê—Ä—Ö—ñ—Ç–µ–∫—Ç—É—Ä–∞ CNN:**  
```
–í—Ö—ñ–¥–Ω–µ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è ‚Üí –ó–≥–æ—Ä—Ç–∫–∞ ‚Üí Pooling ‚Üí –ó–≥–æ—Ä—Ç–∫–∞ ‚Üí Pooling ‚Üí –ü–æ–≤–Ω–æ–∑–≤‚Äô—è–∑–Ω–∏–π —à–∞—Ä ‚Üí –í–∏—Ö—ñ–¥
```

### **2.2. –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–≥–æ—Ä—Ç–∫–æ–≤–æ—ó –Ω–µ–π—Ä–æ–º–µ—Ä–µ–∂—ñ –¥–ª—è —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è –æ–±‚Äô—î–∫—Ç—ñ–≤**  
üìå **–ü—Ä–∏–∫–ª–∞–¥ –ø—Ä–æ—Å—Ç–æ—ó CNN —É TensorFlow:**  
```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞—Ç–∞—Å–µ—Ç—É MNIST (–∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è —Ä—É–∫–æ–ø–∏—Å–Ω–∏—Ö —Ü–∏—Ñ—Ä)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# –ù–æ—Ä–º–∞–ª—ñ–∑–∞—Ü—ñ—è —Ç–∞ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä–æ–∑–º—ñ—Ä—É
X_train = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –º—ñ—Ç–æ–∫ —É one-hot encoding
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# –ü–æ–±—É–¥–æ–≤–∞ CNN-–º–æ–¥–µ–ª—ñ
model = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')  # 10 –∫–ª–∞—Å—ñ–≤ (—Ü–∏—Ñ—Ä–∏ 0-9)
])

# –ö–æ–º–ø—ñ–ª—å–æ–≤—É—î–º–æ –º–æ–¥–µ–ª—å
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# –ù–∞–≤—á–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

üìå **–û—á—ñ–∫—É–≤–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:**  
```
Epoch 5/5
accuracy: 98.5%
```

## **3. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ OpenCV (YOLO) –¥–ª—è —Å–µ–≥–º–µ–Ω—Ç–∞—Ü—ñ—ó –∑–æ–±—Ä–∞–∂–µ–Ω—å (50 —Ö–≤)**  

### **3.1. –©–æ —Ç–∞–∫–µ YOLO?**  
üìå **YOLO (You Only Look Once)** ‚Äì —Ü–µ –æ–¥–Ω–∞ –∑ –Ω–∞–π—à–≤–∏–¥—à–∏—Ö –º–æ–¥–µ–ª–µ–π –≤–∏—è–≤–ª–µ–Ω–Ω—è –æ–±‚Äô—î–∫—Ç—ñ–≤.  
üîπ **–û—Å–Ω–æ–≤–Ω—ñ –ø–µ—Ä–µ–≤–∞–≥–∏:**  
‚úÖ –í–∏—Å–æ–∫–∞ —à–≤–∏–¥–∫—ñ—Å—Ç—å —Ä–æ–±–æ—Ç–∏.  
‚úÖ –í–∏–∑–Ω–∞—á–∞—î **–¥–µ–∫—ñ–ª—å–∫–∞ –æ–±'—î–∫—Ç—ñ–≤** –Ω–∞ –æ–¥–Ω–æ–º—É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ.  
‚úÖ –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –≤ **–±–µ–∑–ø—ñ–ª–æ—Ç–Ω–∏—Ö —Å–∏—Å—Ç–µ–º–∞—Ö, —Ä–æ–∑–≤—ñ–¥—Ü—ñ, –≤—ñ–¥–µ–æ–∞–Ω–∞–ª—ñ—Ç–∏—Ü—ñ**.  

üìå **–ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è YOLO –≤ OpenCV –¥–ª—è —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è –æ–±‚Äô—î–∫—Ç—ñ–≤ —É –≤—ñ–¥–µ–æ–ø–æ—Ç–æ—Ü—ñ:**  

```python
import cv2
import numpy as np

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –Ω–∞—Ç—Ä–µ–Ω–æ–≤–∞–Ω–æ—ó –º–æ–¥–µ–ª—ñ YOLO
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
image = cv2.imread("test_image.jpg")
height, width, channels = image.shape

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è —É —Ñ–æ—Ä–º–∞—Ç YOLO
blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), swapRB=True, crop=False)
net.setInput(blob)
detections = net.forward(output_layers)

# –û–±—Ä–æ–±–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
for detection in detections:
    for obj in detection:
        scores = obj[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]

        if confidence > 0.5:
            center_x, center_y, w, h = (obj[0:4] * np.array([width, height, width, height])).astype("int")
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
cv2.imshow("Detected Objects", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

üìå **–û—á—ñ–∫—É–≤–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:**  
- –ù–∞ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ –±—É–¥–µ **–≤–∏–¥—ñ–ª–µ–Ω–æ –ø—Ä—è–º–æ–∫—É—Ç–Ω–∏–∫–∞–º–∏** –≤–∏—è–≤–ª–µ–Ω—ñ –æ–±‚Äô—î–∫—Ç–∏ (–ª—é–¥–∏, —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç —Ç–æ—â–æ).

```python
import cv2

# –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
image_path = "test_image.jpg"
image = cv2.imread(image_path)

# –ó–∞–ø—É—Å–∫–∞—î–º–æ YOLOv8 –Ω–∞ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ
results = model(image)

# –í—ñ–¥–æ–±—Ä–∞–∂–∞—î–º–æ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è –∑ –¥–µ—Ç–µ–∫—Ç–æ–≤–∞–Ω–∏–º–∏ –æ–±'—î–∫—Ç–∞–º–∏
for result in results:
    result.show()  # –ê–±–æ result.save("output.jpg") –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è
```

‚úÖ YOLO –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –∑–Ω–∞—Ö–æ–¥–∏—Ç—å –æ–±'—î–∫—Ç–∏, –º–∞–ª—é—î —Ä–∞–º–∫–∏ –Ω–∞–≤–∫–æ–ª–æ –Ω–∏—Ö —ñ –¥–æ–¥–∞—î –º—ñ—Ç–∫–∏ –∫–ª–∞—Å—ñ–≤.

### **4. –î–µ—Ç–µ–∫—Ü—ñ—è –æ–±'—î–∫—Ç—ñ–≤ —É —Ä–µ–∞–ª—å–Ω–æ–º—É —á–∞—Å—ñ (–≤—ñ–¥–µ–æ–ø–æ—Ç—ñ–∫)**
üìå **–ó–∞–ø—É—Å–∫–∞—î–º–æ –∫–∞–º–µ—Ä—É —Ç–∞ –≤–∏–∫–æ–Ω—É—î–º–æ –¥–µ—Ç–µ–∫—Ü—ñ—é**
```python
cap = cv2.VideoCapture(0)  # –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤–µ–±–∫–∞–º–µ—Ä–∏

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # –í–∏–∫–æ–Ω—É—î–º–æ –¥–µ—Ç–µ–∫—Ü—ñ—é –æ–±'—î–∫—Ç—ñ–≤
    results = model(frame)

    # –í—ñ–¥–æ–±—Ä–∞–∂–∞—î–º–æ –∫–∞–¥—Ä —ñ–∑ –¥–µ—Ç–µ–∫—Ü—ñ—î—é
    for result in results:
        result.show()

    # –ù–∞—Ç–∏—Å–Ω—ñ—Ç—å "q" –¥–ª—è –≤–∏—Ö–æ–¥—É
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
