## **2. –ü—Ä–∞–∫—Ç–∏—á–Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è OpenCV (40 —Ö–≤)**  

### **2.1. –û—Å–Ω–æ–≤–∏ —Ä–æ–±–æ—Ç–∏ –∑ OpenCV**  
üìå **OpenCV (Open Source Computer Vision Library)** ‚Äì —Ü–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∞ –¥–ª—è –æ–±—Ä–æ–±–∫–∏ –∑–æ–±—Ä–∞–∂–µ–Ω—å —Ç–∞ –∫–æ–º–ø‚Äô—é—Ç–µ—Ä–Ω–æ–≥–æ –∑–æ—Ä—É.  

üîπ **–û—Å–Ω–æ–≤–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó OpenCV:**  
‚úî –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è, –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞ –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω—å.  
‚úî –†–æ–±–æ—Ç–∞ –∑ —Ñ—ñ–ª—å—Ç—Ä–∞–º–∏ (—Ä–æ–∑–º–∏—Ç—Ç—è, –∫–æ–Ω—Ç—Ä–∞—Å—Ç–Ω—ñ—Å—Ç—å, —è—Å–∫—Ä–∞–≤—ñ—Å—Ç—å).  
‚úî –í–∏—è–≤–ª–µ–Ω–Ω—è –∫–æ–Ω—Ç—É—Ä—ñ–≤, —Å–µ–≥–º–µ–Ω—Ç–∞—Ü—ñ—è –æ–±‚Äô—î–∫—Ç—ñ–≤.  

---

### **2.2. –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –æ–±—Ä–æ–±–∫–∞ –∑–æ–±—Ä–∞–∂–µ–Ω—å**  
üìå **–ü—Ä–∏–∫–ª–∞–¥: –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è —Ç–∞ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è –≤ OpenCV**  
```python
import cv2

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
image = cv2.imread("test_image.jpg")

# –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

### **2.3. –ü–æ–ø–µ—Ä–µ–¥–Ω—è –æ–±—Ä–æ–±–∫–∞ –∑–æ–±—Ä–∞–∂–µ–Ω—å**  
üìå **–ó–º—ñ–Ω–∞ —Ä–æ–∑–º—ñ—Ä—É, –ø–µ—Ä–µ–≤–µ–¥–µ–Ω–Ω—è —É –≤—ñ–¥—Ç—ñ–Ω–∫–∏ —Å—ñ—Ä–æ–≥–æ, –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è —Ñ—ñ–ª—å—Ç—Ä—ñ–≤**  
```python
# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —É –≤—ñ–¥—Ç—ñ–Ω–∫–∏ —Å—ñ—Ä–æ–≥–æ
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# –ó–º—ñ–Ω–∞ —Ä–æ–∑–º—ñ—Ä—É
resized = cv2.resize(gray, (128, 128))

# –ó–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è —Ä–æ–∑–º–∏—Ç—Ç—è (Gaussian Blur)
blurred = cv2.GaussianBlur(resized, (5, 5), 0)

# –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
cv2.imshow("Gray Image", gray)
cv2.imshow("Resized Image", resized)
cv2.imshow("Blurred Image", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

---

### **2.4. –í–∏—è–≤–ª–µ–Ω–Ω—è –∫–æ–Ω—Ç—É—Ä—ñ–≤ —Ç–∞ –æ–±‚Äô—î–∫—Ç—ñ–≤**  
üìå **–ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è Canny Edge Detection –¥–ª—è –ø–æ—à—É–∫—É –∫–æ–Ω—Ç—É—Ä—ñ–≤**  
```python
edges = cv2.Canny(gray, 50, 150)

cv2.imshow("Edges", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

üìå **–û—á—ñ–∫—É–≤–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:**  
- –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –∫–æ–Ω—Ç—É—Ä—ñ–≤ –æ–±'—î–∫—Ç—ñ–≤ —É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ.  

---

### **2.5. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∫–∞—Å–∫–∞–¥–Ω–∏—Ö –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ç–æ—Ä—ñ–≤ Haar –¥–ª—è —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è –æ–±‚Äô—î–∫—Ç—ñ–≤**  
üìå **–ü—Ä–∏–∫–ª–∞–¥ –≤–∏—è–≤–ª–µ–Ω–Ω—è –æ–±–ª–∏—á —É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ**  
```python
# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –Ω–∞—Ç—Ä–µ–Ω–æ–≤–∞–Ω–æ–≥–æ –∫–∞—Å–∫–∞–¥—É
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# –í–∏—è–≤–ª–µ–Ω–Ω—è –æ–±–ª–∏—á
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# –ù–∞–Ω–µ—Å–µ–Ω–Ω—è —Ä–∞–º–æ–∫ –Ω–∞ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

cv2.imshow("Detected Faces", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

üìå **–û—á—ñ–∫—É–≤–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:**  
- –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –æ–±–ª–∏—á —ñ–∑ –Ω–∞–Ω–µ—Å–µ–Ω–∏–º–∏ —Ä–∞–º–∫–∞–º–∏.  

---

## **3. –ü—Ä–∞–∫—Ç–∏—á–Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∑–≥–æ—Ä—Ç–∫–æ–≤–∏—Ö –Ω–µ–π—Ä–æ–Ω–Ω–∏—Ö –º–µ—Ä–µ–∂ (CNN) (50 —Ö–≤)**  

### **3.1. –©–æ —Ç–∞–∫–µ –∑–≥–æ—Ä—Ç–∫–æ–≤—ñ –Ω–µ–π—Ä–æ–Ω–Ω—ñ –º–µ—Ä–µ–∂—ñ (CNN)?**  
üìå **CNN (Convolutional Neural Networks)** ‚Äì —Ü–µ –≥–ª–∏–±–æ–∫—ñ –Ω–µ–π—Ä–æ–Ω–Ω—ñ –º–µ—Ä–µ–∂—ñ –¥–ª—è —Ä–æ–±–æ—Ç–∏ —ñ–∑ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è–º–∏.  

üîπ **–û—Å–Ω–æ–≤–Ω—ñ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∏ CNN:**  
‚úî **–ó–≥–æ—Ä—Ç–∫–æ–≤–∏–π —à–∞—Ä (Convolution Layer)** ‚Äì –≤–∏–¥—ñ–ª—è—î –æ–∑–Ω–∞–∫–∏ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è.  
‚úî **Pooling (–°—É–±–¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü—ñ—è)** ‚Äì –∑–º–µ–Ω—à—É—î —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –¥–∞–Ω–∏—Ö.  
‚úî **Fully Connected Layers** ‚Äì –ø—Ä–∏–π–Ω—è—Ç—Ç—è —Ä—ñ—à–µ–Ω—å –Ω–∞ –æ—Å–Ω–æ–≤—ñ –æ–∑–Ω–∞–∫.  

---

### **3.2. –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ CNN –¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó –∑–æ–±—Ä–∞–∂–µ–Ω—å**  
üìå **–ü—Ä–∏–∫–ª–∞–¥ –º–æ–¥–µ–ª—ñ –¥–ª—è —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è —Ä—É–∫–æ–ø–∏—Å–Ω–∏—Ö —Ü–∏—Ñ—Ä (MNIST)**  
```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞—Ç–∞—Å–µ—Ç—É
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω—å
X_train = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ CNN
model = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')  # 10 –∫–ª–∞—Å—ñ–≤ (—Ü–∏—Ñ—Ä–∏ 0-9)
])

# –ö–æ–º–ø—ñ–ª—å–æ–≤—É—î–º–æ –º–æ–¥–µ–ª—å
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# –ù–∞–≤—á–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

üìå **–û—á—ñ–∫—É–≤–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:**  
```
Epoch 5/5
accuracy: 98.5%
```

üöÄ **–ö–ª—é—á–æ–≤–∏–π –º–µ—Å–µ–¥–∂:**  
**–ü–æ—î–¥–Ω–∞–Ω–Ω—è OpenCV —Ç–∞ CNN –¥–æ–∑–≤–æ–ª—è—î —Å—Ç–≤–æ—Ä—é–≤–∞—Ç–∏ –≤–∏—Å–æ–∫–æ—Ç–æ—á–Ω—ñ —Å–∏—Å—Ç–µ–º–∏ —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è –æ–±'—î–∫—Ç—ñ–≤!**
